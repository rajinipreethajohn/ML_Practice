#Linear Regression

```python
#Housing price dataset
#from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Reading the dataset to a pandas dataframe
df = pd.read_csv('/content/housing.csv',  sep='\s+', header = None)
df.head()
```

```python
df.shape
```

```python
# Replace 'column_names' with the list of feature names you want to assign
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

# Assign column names to the DataFrame
df.columns = column_names
```

```python
df.head()
```

```python
df.shape
```

```python
#Our target feature is the last column in the dataset.
#Lets rename it to 'TARGET_PRICE'

df = df.rename(columns={'MEDV' : 'TARGET_PRICE'})
```

```python
df.info()
```

```python
# Split dataeset into train and test datasets
# X will be the independant features
X = df.iloc[:, :13]
X.head()
```

```python
# y will be the target or dependant feature
y = df.iloc[:, -1]
y.head()
```

#Linear Regression


```python
# Relevant pacakages and libraries

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

LR = LinearRegression()
mse = cross_val_score(LR, X, y, scoring = 'neg_mean_squared_error', cv=5)
```

```python
mse
```

```python
mean_mse= np.mean(mse)
mean_mse
```

#Ridge Regression

```python
# Relevant pacakages and libraries

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
params = {'alpha':[1e-15,1e-10,1e-8, 1e-3, 1e-2,1,5,10,20]}

ridge_regressor = GridSearchCV(ridge, params, scoring ='neg_mean_squared_error', cv=5)
ridge_regressor.fit(X, y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

#Lasso Regression


```python
# Relevant pacakages and libraries

from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()
params = {'alpha':[1e-15,1e-10,1e-8, 1e-3, 1e-2,1,5,10,20]}

lasso_regressor = GridSearchCV(ridge, params, scoring ='neg_mean_squared_error', cv=5)
lasso_regressor.fit(X, y)
```
