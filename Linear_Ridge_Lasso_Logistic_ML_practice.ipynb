#Linear Regression

```python
#Housing price dataset
#from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Reading the dataset to a pandas dataframe
df = pd.read_csv('/content/housing.csv',  sep='\s+', header = None)
df.head()
```

```python
df.shape
```

```python
# Replace 'column_names' with the list of feature names you want to assign
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

# Assign column names to the DataFrame
df.columns = column_names
```

```python
df.head()
```

```python
df.shape
```

```python
#Our target feature is the last column in the dataset.
#Lets rename it to 'TARGET_PRICE'

df = df.rename(columns={'MEDV' : 'TARGET_PRICE'})
```

```python
df.info()
```

#Linear Regression


```python
X = df.iloc[:, :13]
X.head()
```

```python
y = df.iloc[:, -1]
y.head()
```

```python
#Train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.33, random_state =42)
```

```python
# Relevant pacakages and libraries

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

LR = LinearRegression()
mse = cross_val_score(LR, X_train, y_train, scoring = 'neg_mean_squared_error', cv=5)
```

```python
mse
```

```python
mean_mse= np.mean(mse)
mean_mse
```

#Ridge Regression

```python
# Relevant pacakages and libraries

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
params = {'alpha':[1e-15,1e-10,1e-8, 1e-3, 1e-2,1,5,10,20]}

ridge_regressor = GridSearchCV(ridge, params, scoring ='neg_mean_squared_error', cv=5)
ridge_regressor.fit(X, y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

#Lasso Regression


```python
# Relevant pacakages and libraries

from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()
params = {'alpha':[1e-15,1e-10,1e-8, 1e-3, 1e-2,1,5,10,20]}

lasso_regressor = GridSearchCV(ridge, params, scoring ='neg_mean_squared_error', cv=5)
lasso_regressor.fit(X, y)
```

#Logistic Resgression on Breast Cancer Dataset


```python
#Load breast cancer dataset from sklearn datsets
from sklearn.datasets import load_breast_cancer

df= load_breast_cancer()
type(df)
```

```python
# Lets look at the independant features (X) =
X = pd.DataFrame(df['data'], columns =df['feature_names'])
X.head()
```

```python
#Lets look at the dependant/ target variable
y = pd.DataFrame(df['target'], columns=['Target'])
y.head()
```

```python
#Checking if the y dataset is balanced or not
y.value_counts()
```

```python
#Train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.33, random_state =42)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
```

```python
params = [{'C': [1,5,10]}, {'max_iter' : [100, 150]}]
```

```python
model1 = LogisticRegression(C=100, max_iter=100)
```

```python
model = GridSearchCV(model1,  param_grid=params, scoring= 'f1', cv=5)
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred= model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_pred, y_test))
```

```python
print("Accuracy Score :" , accuracy_score(y_pred, y_test))
```

```python
print(classification_report(y_pred, y_test))
```
